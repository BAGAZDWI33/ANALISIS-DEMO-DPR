```python
# =========================================
# ANALISIS SENTIMEN DATA TWITTER
# =========================================

import pandas as pd
import re
import matplotlib.pyplot as plt

# 1. BACA DATA
df = pd.read_csv("/content/tweets-data/demodpr.csv")

# 2. PREPROCESSING TEKS
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)  # hapus URL
    text = re.sub(r"@\w+", "", text)            # hapus mention
    text = re.sub(r"#", "", text)               # hapus hashtag
    text = re.sub(r"[^a-zA-Z\s]", "", text)     # hapus angka & simbol
    return text.strip()

df['clean_text'] = df['full_text'].apply(clean_text)

# 3. BUAT KAMUS SENTIMEN SEDERHANA
positif_words = ["bagus", "baik", "setuju", "hebat", "mendukung", "suka", "puas", "mantap", "keren", "terbaik"]
negatif_words = ["jelek", "buruk", "korup", "marah", "tolol", "anjing", "kecewa", "benci", "tolak", "gagal"]

def get_sentiment(text):
    pos_count = sum(word in text for word in positif_words)
    neg_count = sum(word in text for word in negatif_words)
    if pos_count > neg_count:
        return "Positif"
    elif neg_count > pos_count:
        return "Negatif"
    else:
        return "Netral"

df['sentimen'] = df['clean_text'].apply(get_sentiment)

# 4. PROPORSI SENTIMEN
sentiment_count = df['sentimen'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(sentiment_count, labels=sentiment_count.index, autopct='%1.1f%%', startangle=140)
plt.title("Proporsi Sentimen Tweet")
plt.show()

# 5. TIMELINE SENTIMEN
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
timeline = df.groupby([df['created_at'].dt.floor('H'), 'sentimen']).size().unstack(fill_value=0)

plt.figure(figsize=(12,5))
timeline.plot(kind='line')
plt.title("Timeline Jumlah Tweet per Jam berdasarkan Sentimen")
plt.xlabel("Waktu")
plt.ylabel("Jumlah Tweet")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. ENGAGEMENT RATA-RATA PER SENTIMEN
engagement = df.groupby('sentimen')[['favorite_count','retweet_count','reply_count']].mean()

engagement.plot(kind='bar', figsize=(8,5))
plt.title("Rata-rata Engagement per Sentimen")
plt.ylabel("Rata-rata")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 7. SIMPAN HASIL KE FILE CSV
df.to_csv("hasil_analisis_sentimen.csv", index=False)

print("Analisis selesai! File 'hasil_analisis_sentimen.csv' sudah disimpan.")
```
